我们可以进一步通过pruning来找出目前的DS迭代步实现在计算上需要优化的地方。下面将以PDS系列问题为例进行分析。

```python
%load_ext line_profiler
```

```python
ds = DualSimplexSolver()
fnames = [fname for fname in sorted(glob.glob('netlib/*.SIF')) if 'PDS' in fname]
%lprun -f ds._step -T prun_step test(ds.solve,model_fnames=fnames,random_seed=32)
```

```python
import pandas as pd
def eval_(s):
    return eval(s) if s else 0

with open('prun_step','r+') as f:
    lines = f.readlines()
result = []
for line in lines[8:]:
    result_temp = [line[:6].strip(),line[6:16].strip(),line[16:29].strip(),line[29:38].strip(),line[38:47].strip()]
    result_temp = [eval_(item) for item in result_temp]
    result_temp += [line[47:].strip()]
    result += [result_temp]
    
## 按代码顺序排列
result = pd.DataFrame(result,columns=['row num','count','time','unit time','%','code'])
result['cum %'] = (result['time'].cumsum() / result['time'].sum() * 10000).astype(int) / 100
result = result[['row num','count','time','unit time','%','cum %','code']]
# print(result.to_string())

## 按计算开销排列
result = result.sort_values(['time'],ascending=False)
result['cum %'] = (result['time'].cumsum() / result['time'].sum() * 10000).astype(int) / 100
result = result[['row num','count','time','unit time','%','cum %','code']]
print(result.head(30).to_string())
```

可以看到在PDS系列问题上，线性方程求解、BFRT、DSE pricing（包括权重更新）分别占50%、10%和15%的计算开销，余下20%多的计算开销主要来自于其他一些向量相关的计算和内存管理。最后这部分计算的效率较难在原生Python上进行优化；用其他更适合做计算的语言实现应该可以有一定的提升。
